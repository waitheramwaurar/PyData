### Masks, Comparisons and Boolean Operators in Numpy

```python
import numpy as np

weights = np.array([52, 43, 61, 30, 72, 51, 28, 64])

# How do we filter this array of weights to give us 'obese' patients? Using masks

weights
```

```python
obese_mask = weights > 60
# Mask is a special type of array where the values are Boolean
obese_mask
```

```python
# Applying the mask

obese_patients = weights[obese_mask]
obese_patients
```

```python
# Moderate weight mask

moderate_weight_mask = (weights >= 40) & (weights <= 60)
moderate_weight_mask
```

```python
#  Inverting the obese mask
not_obese_mask = ~(weights > 60)
not_obese_mask
```

#### Filtering dataframes based on conditions

```python
sales_df.head()
```

```python
# Filter the dataframe for all phones above 15,000
sales_df[sales_df['Selling Price'] >15000]
```

```python
# Using loc
sales_df.loc[sales_df["Selling Price"]>15000]
```

```python
# Use loc to view mobile phones is more than 8GB
sales_df.loc[sales_df['Memory'] == '8 GB']
```

```python
# If you want to apply more than one condition
# Xiaomi phones that have 128 GB storage

sales_df[(sales_df['Brands'] == 'Xiaomi') | (sales_df['Storage'] == '128 GB')].head()
```

```python
# All phones that have 4GB ram and cost more than 10k

sales_df.loc[(sales_df['Memory'] == '4 GB') & (sales_df['Selling Price'] > 10000)]
```

```python
# Filter the dataframe for all Apple and Samsung phones which have no discount

sales_df.loc[(sales_df['Brands'].isin(['Apple', 'Samsung', 'Lenovo'])) & (sales_df['Discount'] == 0)]
```

#### Grouping and aggregation

```python
# groupby function - split-apply-combine
# split df based on brand, apply the aggregate function, combine into an ouput
# Can only give an output when it is used by an aggregate function e.g average, sum, max etc.

# Group based on brands
# df.groupby(column_where_split_is_happening)[column_where_the_aggregate_fn_is_being_applied].aggregate_fn
# df.groupby('split')['apply'].agg()

sales_df.groupby('Brands')['Selling Price'].mean()
```

```python
# What if we want to sort?
sales_df.groupby('Brands')['Selling Price'].mean().sort_values(ascending=False)
```

```python
# Which brand has been given the highest average discount

sales_df.groupby('Brands')['Discount'].mean().idxmax()
```

```python
# Which brand did we sell the most
sales_df.groupby('Brands')['Selling Price'].count().idxmax()
```

```python
# Which brand has the highest number of models
sales_df.groupby('Brands')['Models'].nunique().sort_values(ascending=False)
```

```python
# You can apply your own customized function using lambda function

# Identify the brand with the larges

sales_df.groupby('Brands')['Selling Price'].agg(lambda x: x.max() - x.min()).sort_values()
```
