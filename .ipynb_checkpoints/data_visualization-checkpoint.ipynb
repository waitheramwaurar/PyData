## Visualization with Python Matplotlib

- Bar plots: vertical, horizontal, stacked, grouped
- Customizations
- Scatter plots and color bars

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
drinks = pd.read_csv('drinks.csv')
drinks.head()
```

### Bar plots - visualize the counts of categorical features/variables

```python
# Plot beer servings per country

plt.bar(drinks['country'], drinks['beer_servings'], color='blue', edgecolor='black')
plt.show()
```

```python
# Sorting values to have a better visualization
top_10 = drinks.sort_values(by='beer_servings', ascending=False).head(10)
top_10
```

```python
plt.bar(top_10['country'], top_10['beer_servings'], color='blue', edgecolor='black')
plt.title('Beer Servings per Person in Top 10 Countries')
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.ylabel('Average beer servings per person')
plt.show()
```

```python
# Advanced customization - adding annotations

plt.bar(top_10['country'], top_10['beer_servings'], color='blue', edgecolor='black')
plt.title('Beer Servings per Person in Top 10 Countries')
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.ylabel('Average beer servings per person')
# make y-axis invisible
plt.gca().get_yaxis().set_visible(False)

# plt.text() - define for it an x and y coordinate
# enumerate gives value and index
for x, y in enumerate(top_10['beer_servings']):
    plt.text(x, y+5, str(y), ha='center')
# ha - orientation
plt.show()
```

### Grouped bar chart

```python
bar_width = 0.5
x = np.arange(len(top_10['country']))
# where to plot the bars
plt.bar(x-bar_width/2, top_10['beer_servings'], bar_width, label='Beer', color='blue')
plt.bar(x+bar_width/2, top_10['wine_servings'], bar_width, label='Wine', color='gray')

plt.xticks(x, top_10['country'], rotation=90)
plt.title('Beer & Wine Servings per Person in Top 10 Countries')

for x, y in enumerate(top_10['beer_servings']):
    plt.text(x-0.2, y+3, str(y), ha='center')

for x, y in enumerate(top_10['wine_servings']):
    plt.text(x+0.2, y+3, str(y), ha='center')
plt.show()
```

### Assignment: Plot a visualization to compare the beer, spirits and wine servings of Uganda, Kenya and Tanzania

```python
# Stacked bar

plt.figure(figsize=(10,6))
plt.bar(top_10['country'], top_10['beer_servings'], color='brown')
plt.bar(top_10['country'], top_10['spirit_servings'], bottom=top_10['beer_servings'], color='purple')
plt.bar(top_10['country'], top_10['wine_servings'], bottom=top_10['beer_servings']+top_10['spirit_servings'], color='red')
plt.xticks(rotation=90)
plt.show()
```

### Plotting multiple plots on the same figure

```python

fig, axs = plt.subplots(1,3, figsize=(10,6)) # give me one figure (1 row, 3 columns) and give me the axes

axs[0].bar(top_10['country'], top_10['beer_servings'])
axs[1].bar(top_10['country'], top_10['wine_servings'])
axs[2].bar(top_10['country'], top_10['spirit_servings'])
```

```python
# Having multiple rows 
fig, axs = plt.subplots(2,2, figsize=(12,5))
axs[0,0].bar(top_10['country'], top_10['beer_servings'])
axs[0,1].bar(top_10['country'], top_10['wine_servings'])
axs[1,0].bar(top_10['country'], top_10['spirit_servings'])
axs[1,1].bar(top_10['country'], top_10['total_litres_of_pure_alcohol'])
```
