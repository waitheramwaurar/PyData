### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')['age'].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
# Total people per occupation
total_people = users.groupby('occupation')['gender'].size()

total_males = users.loc[users['gender'] == 'M'].groupby('occupation').size()

male_ratio = (total_males/total_people).sort_values(ascending=False)
male_ratio
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
wages = pd.DataFrame({
    'min_age': users.groupby('occupation')['age'].min(),
    'max_age': users.groupby('occupation')['age'].max()
})
wages
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
mean_age = users.groupby(['occupation', 'gender'])['age'].mean().round(2).unstack(fill_value=0)
mean_age
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender_counts = users.groupby(['occupation', 'gender']).size()

total_counts = users.groupby('occupation').size()

gender_percent = (gender_counts / total_counts) * 100

gender_percent = gender_percent.unstack(fill_value=0).round(1)
gender_percent
```
